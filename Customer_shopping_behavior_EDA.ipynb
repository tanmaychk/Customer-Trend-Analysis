```python
#load into dataframe
import pandas as pd
df  =pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
#metadata
df.info()
```

```python
#stat summary
df.describe()
```

```python
#find null values
df.isnull().sum()
```

```python
# replace null review ratings by median for each category over means as outliers might skew data
# find median rating over each category and replace it with null value.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#find null values
df.isnull().sum()
```

```python
#snake case column names for consistent references
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
#group customer into age groups based on the age columns
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create a column for purchase frequency days from given frequncy_of_purchase column:
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly': 14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
# check for column redundancy of 'discount_applied' and 'promo_code_used' columns
df[['discount_applied','promo_code_used']].head(10)
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# dropping column promo code used
df = df.drop('promo_code_used',axis=1)
```

```python
df=df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```
